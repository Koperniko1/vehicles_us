```python
import pandas as pd
```

```python
vehicles = pd.read_csv('C:/Users/Main/Desktop/Proyecto/vehicles_us/vehicles_us.csv')
```

```python
vehicles.info()
```

```python
unique_values = {}

for column in vehicles.columns:
    unique_vals = vehicles[column].dropna().unique()
    if vehicles[column].dtype in ['float64', 'int64']:
        unique_vals = sorted(unique_vals)
    unique_values[column] = unique_vals

# Imprimir los valores Ãºnicos
for column, values in unique_values.items():
    print(f"Valores Ãºnicos de la columna '{column}':")
    print(values)
    print()
```
